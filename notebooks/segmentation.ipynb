```python
# 1) 필수 라이브러리 로드
from ultralytics import YOLO
from tqdm import tqdm
import cv2
import os
```

```python
# 2) 경로 설정
PROJECT_ROOT = os.path.abspath('..')           # notebooks/ 기준
VIDEO_PATH   = os.path.join(PROJECT_ROOT, 'input', 'input.mov')
OUT_DIR      = os.path.join(PROJECT_ROOT, 'outputs', 'runs', 'segment', 'predict', 'input.mov')
FRAMES_DIR   = os.path.join(PROJECT_ROOT, 'outputs', 'runs', 'segment', 'predict', 'frames')

os.makedirs(FRAMES_DIR, exist_ok=True)
```

```python
# 3) 모델 불러오기
model = YOLO('yolov8s-seg.pt')
```

```python
# 4) 동영상 바로 세그멘테이션 수행 (Ultralytics API 사용)
results = model.predict(
    source=VIDEO_PATH,
    save=True,
    project=os.path.join(PROJECT_ROOT, 'outputs', 'runs', 'segment'),
    name='predict',
    conf=0.25,        # 신뢰도 임계값
    show=False,       # 결과 윈도우를 띄우지 않음
)
```
